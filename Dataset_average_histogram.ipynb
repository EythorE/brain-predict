```python
import os
import numpy as np
import pandas as pd
import nibabel as nib #Read / write access to some common neuroimaging file formats
import matplotlib.pyplot as plt

data_dir = './data'
```

# IXI dataset

```python
csv_IXI = 'IXI_train.csv'
df_IXI = pd.read_csv(os.path.join(data_dir, csv_IXI))
filename_IXI = np.array(df_IXI['file'])
print("Number of brains: ", len(filename_IXI))
one_IXI = np.array(nib.load( os.path.join(data_dir, filename_IXI[42]) ).get_data())
shape = one_IXI.shape
print(one_IXI.dtype, one_IXI.itemsize)
print(one_IXI.size)
print("Brain image shape: ", shape)
dataset_size = len(filename_IXI)*one_IXI.size*one_IXI.itemsize
print("Size =", np.round(dataset_size/1024**3, 2), "GB")
```

### Histogram of brain images mean value

```python
IXI_means = []
sum_IXI = np.zeros_like(one_IXI) # Used later
for brain in filename_IXI:
    image = nib.load(os.path.join(data_dir, brain)).get_data()
    IXI_means.append(np.average(image))
    sum_IXI = sum_IXI + np.array(image)
```

```python
a = plt.hist(IXI_means, bins=10)
print(a[0])
```

```python
outliers = np.where(np.array(IXI_means)>1000)
print(outliers)
```

```python
dimension = 0
s = 100
plt.figure(figsize=(20, 20))

plt.subplot(2,3,1)
plt.imshow(np.take(nib.load(os.path.join(data_dir, filename_IXI[30])).get_data(), s, dimension))
plt.title("mean: "+str(IXI_means[30]))
plt.subplot(2,3,2)
plt.imshow(np.take(nib.load(os.path.join(data_dir, filename_IXI[336])).get_data(), s, dimension))
plt.title("mean: "+str(IXI_means[336]))
plt.subplot(2,3,3)
plt.imshow(np.take(nib.load(os.path.join(data_dir, filename_IXI[357])).get_data(), s, dimension))
plt.title("mean: "+str(IXI_means[357]))

for i in range(4, 7):
    index = np.random.randint(0,len(filename_IXI))
    plt.subplot(2,3,i)
    plt.imshow(np.take(nib.load(os.path.join(data_dir, filename_IXI[index])).get_data(), s, dimension))
    plt.title("mean: "+str(IXI_means[index]))
plt.show()
```

# Average brains

```python
average_all = (sum_IXI+sum_OASIS+sum_openfmri)/(len(filename_IXI)+len(filename_OASIS)+len(filename_openfmri))
average_IXI = sum_IXI/len(filename_IXI)
average_OASIS = sum_OASIS/len(filename_OASIS)
average_openfmri = sum_openfmri/len(filename_openfmri)
```

```python
dimension = 0
s = 100
plt.figure(figsize=(20, 20))

plt.subplot(2,3,1)
plt.imshow(np.take(average_OASIS, s, dimension))
plt.title("OASIS")
plt.subplot(2,3,2)
plt.imshow(np.take(average_IXI, s, dimension))
plt.title("IXI")
plt.subplot(2,3,3)
plt.imshow(np.take(average_openfmri, s, dimension))
plt.title("OpenfMRI")

plt.subplot(2,3,5)
plt.imshow(np.take(average_all, s, dimension))
plt.title("Combined dataset average brain")
```
