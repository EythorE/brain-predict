```python
import os
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('oasis.csv')
df.head(5)
```

```python
print(df['Subject'].is_unique)
df.rename(columns={'M/F':'sex'}, inplace=True)
df.head(5)
```

```python
brain_dir = './OASIS3/'
error_brain_dir = './oasis_missing_data/'

brain_files = np.array([f for f in os.listdir(brain_dir)])
num_images = len(brain_files)
print(num_images)
print(brain_files[:4])
```

```python
paths = []
filenames = []
# r=root, d=directories, f = files
for r, d, f in os.walk(brain_dir):
    for file in f:
        if '.nii.gz' in file:
            paths.append(os.path.join(r, file))
            filenames.append(file)

for f,p in zip(filenames[:5], paths[:5]):
    print(f)
    print(p)
```

```python
filename_id = [fname[4:12] for fname in filenames]
filename_id[:5]
```

```python
sex = [df.loc[df['Subject'] == subj].sex.values[0].lower() for subj in filename_id]
for s in sex:
    if not (s == 'f' or s == 'm'):
        print('error')
```

```python
print(len(filename_id))
print(len(filenames))
print(len(paths))
print(len(sex))
```

### Búum til nýja csv skrá

```python
d = {"subject": filename_id, "file":filenames, "path": paths, "sex": sex}
new_df = pd.DataFrame(data=d)
new_df.head(5)
```

```python
new_df.to_csv("OASIS_orig_path_labels.csv", sep=',')
```
