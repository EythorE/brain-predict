```python
import numpy as np
import pandas as pd
df = pd.read_csv('data/IXI/IXI.csv')
df.head(10)
```

```python
subject_dir = np.load('subject.npy')
print(subject_dir[:10])
print('...')
print(subject_dir[-10:])
```

```python
subject_no = np.zeros(subject_dir.shape, dtype=np.int)
for i,subj in enumerate(subject_dir):
    subject_no[i] = subj[3:6]
```

```python
print(subject_no[:10])
print('...')
print(subject_no[-10:])
```

```python
df.columns
```

```python
errors_subj = []
errors_index = []

subject_age = np.zeros(len(subject_no))
subject_gender = np.zeros(len(subject_no), dtype='U1')
for i, subj in enumerate(subject_no):
    if(len(df.loc[df['IXI_ID'] == subj]) != 1):
        print('Error: subject',subj, 'entries =', len(df.loc[df['IXI_ID'] == subj]))
        subject_age[i] = None
        subject_gender[i] = ''
        errors_subj.append(subj)
        errors_index.append(i)
        continue
    subject_age[i] = df.loc[df['IXI_ID'] == subj].AGE.values[0]
    if(df[df.columns[1]].loc[df['IXI_ID'] == subj].values[0] == 1):
        subject_gender[i] = 'm'
    elif(df[df.columns[1]].loc[df['IXI_ID'] == subj].values[0] == 2):
        subject_gender[i] = 'f'
    else:
        print('error gender unknown')
```

```python
df.loc[df['IXI_ID'].isin(errors_subj)]
```

Sjáum að ef það eru tvær línur fyrir subject þá er aldur og kyn oft skráð það sama fyrir bæði, notum þau gildi ef svo er (mest tvær línur per subject)

```python
for i, subj in enumerate(errors_subj):
    if(len(df.loc[df['IXI_ID'] == subj]) == 2):
        
        print('Subject', subj, 'old age =',subject_age[errors_index[i]])
        if(df.loc[df['IXI_ID'] == subj].AGE.values[0] == df.loc[df['IXI_ID'] == subj].AGE.values[1]):
            subject_age[errors_index[i]] = df.loc[df['IXI_ID'] == subj].AGE.values[0]
        print('Subject', subj, 'new age =',subject_age[errors_index[i]])
        
        print('Subject', subj, 'old gender =',subject_gender[errors_index[i]])
        if(df[df.columns[1]].loc[df['IXI_ID'] == subj].values[0] == df[df.columns[1]].loc[df['IXI_ID'] == subj].values[1]):
            if(df[df.columns[1]].loc[df['IXI_ID'] == subj].values[0] == 1):
                subject_gender[errors_index[i]] = 'm'
            elif(df[df.columns[1]].loc[df['IXI_ID'] == subj].values[0] == 2):
                subject_gender[errors_index[i]] = 'f'
        print('Subject', subj, 'new gender =',subject_gender[errors_index[i]])
```

```python
del_subj = np.where(np.logical_or(np.isnan(subject_age), subject_gender==''))
print(subject_no[del_subj])
```

```python
brains = np.load('brains.npy')
```

```python
print(len(brains))
print(len(subject_no))
print(len(subject_age))
print(len(subject_gender))
```

```python
brains = np.delete(brains, del_subj, axis=0)
subject_no = np.delete(subject_no, del_subj, axis=0)
subject_age = np.delete(subject_age, del_subj, axis=0)
subject_gender = np.delete(subject_gender, del_subj, axis=0)
```

```python
print(len(brains))
print(len(subject_no))
print(len(subject_age))
print(len(subject_gender))
```

```python
rand_indicies = np.random.permutation(np.arange(len(brains)))

np.save('IXI_cleaned_train/subject_no.npy', subject_no[rand_indicies[0:500]])
np.save('IXI_cleaned_train/subject_age.npy', subject_age[rand_indicies[0:500]])
np.save('IXI_cleaned_train/subject_gender.npy', subject_gender[rand_indicies[0:500]])
np.save('IXI_cleaned_train/subject_brains.npy', brains[rand_indicies[0:500]])

np.save('IXI_cleaned_test/subject_no.npy', subject_no[rand_indicies[500:]])
np.save('IXI_cleaned_test/subject_age.npy', subject_age[rand_indicies[500:]])
np.save('IXI_cleaned_test/subject_gender.npy', subject_gender[rand_indicies[500:]])
np.save('IXI_cleaned_test/subject_brains.npy', brains[rand_indicies[500:]])
```
