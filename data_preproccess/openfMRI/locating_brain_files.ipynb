```python
import os
import glob
import pandas as pd
import numpy as np
```

```python
path = './openfmri'
t1wglob =  '/sub*/anat*/*T1w.nii.gz'
dsglop='/ds*'
```

```python
df = pd.read_csv('filtered_data.csv')
df.head(5)
```

```python
files = glob.glob(path+dsglop+t1wglob)
```

```python
len(files)
```

```python
files[:10]
```

```python
os.getcwd()
```

```python
dataset_index = []
notfound_index = []

new_name = []
path = []
age = []
sex = []

# Iterate over all labeled data in dataset
for index, row in df.iterrows():
    # find corrisponding brain file
    duplicates = 0
    for file in files:
        if (row['dataset'] in file) and (row['id']+'/' in file):
            duplicates += 1
            if duplicates == 1:
                dataset_index.append(index)
            path.append(file)
            
            new_name.append(row['dataset']+'_'+os.path.basename(file))
            age.append(row['age'])
            sex.append(row['sex'])
    # print duplicate if duplicate images
    if duplicates > 1:
        for i in range(duplicates):
            print(row['dataset'],row['id'], path[-i-1])
    if duplicates == 0:
        notfound_index.append(index)
```

```python
print(len(files), len(new_name), len(notfound_index))
```

```python
index, counts = np.unique(new_name, return_counts=True)
np.count_nonzero(counts > 1)
```

```python
new_df = pd.DataFrame(data={"new_name": new_name,"path": path , "age": age, "sex": sex})
new_df.head(10)
```

```python
new_df.to_csv('available_data.csv', index=False)
```

```python
with open('preprocess.txt', 'w') as f:
    for path, new_path in zip(new_df['path'], new_df['new_name']):
        f.write("%s ./tmp/%s\n" % (path, new_path))
```
