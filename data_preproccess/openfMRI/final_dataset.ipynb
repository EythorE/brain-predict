```python
import os
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('available_data.csv')
df.head(5)
```

```python
path = 'openfmri_brains'
files = os.listdir(path)
```

```python
df.set_index("new_name", drop=True, append=False, inplace=True, verify_integrity=True)
df.head(5)
```

```python
df.loc["ds000009_sub-05_T1w.nii.gz"].age
```

```python
subject_id = [] #openfmri####
data_path = [] #./openfmri/...
age = []
sex = []
for index, file in enumerate(files):
    subject_id.append("openfmri"+str(index).zfill(4))
    data_path.append("./openfmri/"+file)
    age.append(df.loc[file].age)
    sex.append(df.loc[file].sex)
```

```python
new_df = pd.DataFrame(data={"subject": subject_id, "file": data_path, "age": age, "sex": sex})
new_df.head(30)
```

```python
new_df.to_csv("openfmri_labels.csv", sep=',', index=False)
```
